```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Example usage:
pdf_path = '/content/Sure trust.pdf'
pdf_text = extract_text_from_pdf(pdf_path)
```

```python
from llama_index.core import Document

# Create a Document instance from the extracted text
document = Document(text=pdf_text)

# If you have multiple documents, create a list of Document instances
documents = [document]
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts import PromptTemplate
```

```python
system_prompt="""
you are a Q/A assistant . your goal is to answer question as accurately as possible based on the intructions and context provided"""
query_wrapper_prompt = PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
import torch
from transformers import BitsAndBytesConfig
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)
llm = HuggingFaceLLM(
    context_window = 4096,
    max_new_tokens = 256,
    generate_kwargs = {"temperature":0.0,"do_sample": False},
    system_prompt = system_prompt,
    query_wrapper_prompt = query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    model_kwargs={"torch_dtype":torch.float16,"quantization_config": quantization_config}
)
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings

Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-mpnet-base-v2"
)
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size = 1024,
    llm = llm,
    embed_model = Settings.embed_model
)
```

```python
service_context
```

```python
index = VectorStoreIndex.from_documents(documents, service_context=service_context)
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("Who are the Collaborators of Sure trust?")
```
